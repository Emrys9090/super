```python
# Group by "State or Province" and sum the Shipping Cost
top_states = df.groupby("State or Province")["Shipping Cost"].sum()
```

```python
# Sort in descending order and get the top 5 states
top_5_states = top_states.sort_values(ascending=False).head(5)
```

```python
print(top_5_states)
```

```python
# Group by "State or Province" and sum the shipping cost
top_states = df.groupby("State or Province")["Shipping Cost"].sum().nlargest(5)
```

```python
# Plot the data
plt.figure(figsize=(10, 6))
top_states.plot(kind='bar', color='skyblue', edgecolor='black')
```
